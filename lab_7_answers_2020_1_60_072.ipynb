```python
#Average Educational Number by Education Level
data = {
    'education': ['11th', 'HS-grad', 'Assoc-acdm', 'Some-college', '10th',
                  'Assoc-acdm', 'HS-grad', 'HS-grad', 'HS-grad', 'HS-grad'],
    'educational-num': [7, 9, 12, 10, 6, 12, 9, 9, 9, 9]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.barplot(x='education', y='educational-num', data=df, palette='Set2')
plt.title('Average Educational Number by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Educational Number')
plt.show()

```

PIE-CHART

```python
data = {'income': ['<=50K', '<=50K', '>50K', '>50K', '<=50K', '<=50K', '>50K', '<=50K', '<=50K', '>50K']}
df = pd.DataFrame(data)
df['income'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title('Income Distribution')
plt.ylabel('')
plt.show()
```

COUNT-PLOT

```python
data = {
    'education': ['11th', 'HS-grad', 'Assoc-acdm', 'Some-college', '10th',
                  'Assoc-acdm', 'HS-grad', 'HS-grad', 'HS-grad', 'HS-grad'],
    'income': ['<=50K', '<=50K', '>50K', '>50K', '<=50K',
               '<=50K', '>50K', '<=50K', '<=50K', '>50K']
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.countplot(x='education', hue='income', data=df)
plt.title('Education vs Income')
plt.xlabel('Education Level')
plt.ylabel('Count')

plt.show()

```

HISTOGRAM

DOT-PLOT

```python

data = {'occupation':['Machine-op-inspct', 'Farming-fishing', 'Protective-serv',
                   'Machine-op-inspct', 'Other-service', 'Tech-support',
                   'Machine-op-inspct', 'Adm-clerical', 'Adm-clerical', 'Exec-managerial'],
                   'gender': ['Male', 'Male', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']}

df = pd.DataFrame(data)
count_data = df.groupby(['gender', 'occupation']).size().reset_index(name='count')
plt.figure(figsize=(10, 6))
for occupation in count_data['occupation'].unique():
    occupation_data = count_data[count_data['occupation'] == occupation]
    plt.scatter(occupation_data['gender'], occupation_data['occupation'],
                s=occupation_data['count']*30, label=occupation, alpha=0.6)
plt.title('Dot Plot for Gender and Occupation')
plt.xlabel('Gender')
plt.ylabel('Occupation')
plt.xticks(rotation=0)
plt.legend(title='Occupation', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

```

line-plot

```python
data = {
    'workclass': ['Private', 'Private', 'Local-gov', 'Private', 'Private', 'Private', 'Private', 'Private', 'Private', 'Self-emp-inc'],
    'income': ['<=50K', '<=50K', '>50K', '>50K', '<=50K', '<=50K', '>50K', '<=50K', '<=50K', '>50K']
}

df = pd.DataFrame(data)
count_data = df.groupby(['workclass', 'income']).size().reset_index(name='count')
pivot_data = count_data.pivot(index='workclass', columns='income', values='count').fillna(0)
plt.figure(figsize=(10, 6))
pivot_data.plot(kind='line', marker='o', linewidth=2)
plt.title('Line Plot of Workclass and Income')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Income')
plt.tight_layout()
plt.show()
```
