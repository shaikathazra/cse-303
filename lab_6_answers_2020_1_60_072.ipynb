```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
region_deaths = df.groupby('WHO Region')['Deaths'].sum()
```

```python
region_deaths
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=region_deaths.index, y=region_deaths.values)
plt.title('Total Deaths by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Total Deaths')

plt.show()
```

```python
countries = ['Bangladesh', 'India', 'Pakistan']
filtered_df = df[df['Country/Region'].isin(countries)]

plt.figure(figsize=(8, 8))
plt.pie(filtered_df['Deaths'], labels=filtered_df['Country/Region'], autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])

plt.title('COVID-19 Deaths in Selected Countries')
plt.show()
```

##Start Plotting


```python
#Bar chat for who reagon death vs normal_region_deaths death
normal_region_deaths = df.groupby('Country/Region')['Deaths'].sum()
#normal_region_deaths

southeast_asian_countries = ['Myanmar', 'Indonesia', 'Laos', 'Malaysia','Philippines', 'Singapore']
filtered_df = df[df['Country/Region'].isin(southeast_asian_countries)]


plt.figure(figsize=(8, 8))
plt.pie(filtered_df['Deaths'], labels=filtered_df['Country/Region'], autopct='%6.1f%%', startangle=140)

plt.title('COVID-19 Deaths in Selected Countries')
plt.show()




```

```python
region_newcase = df.groupby('WHO Region')['New cases'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=region_newcase.index, y=region_newcase.values)
plt.title('Total new case by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Total newcase')

plt.show()
```

```python
region_newcase = df.groupby('Country/Region')['Recovered'].sum()

plt.figure(figsize=(40, 10))
sns.scatterplot(x=region_newcase.index, y=region_newcase.values)
plt.title('Total new case by WHO Region')
plt.xlabel('WHO Region')
plt.xticks(rotation=90)
plt.ylabel('Total newcase')

plt.show()
```

```python
countries = ['Bangladesh', 'India', 'Pakistan']
filtered_df = df[df['Country/Region'].isin(countries)]

plt.figure(figsize=(8, 6))
plt.bar(filtered_df['Country/Region'], filtered_df['Deaths'], color=['#ff9999', '#66b3ff', '#99ff99'])

plt.xlabel('Country/Region')
plt.ylabel('Number of Deaths')
plt.title('COVID-19 Deaths in Selected Countries')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
region_newcase = df.groupby('Country/Region')['Recovered'].sum()


plt.figure(figsize=(40, 10))
sns.lineplot(x=region_newcase.index, y=region_newcase.values, marker='o', linestyle='-', color='blue')

plt.title('Total Recovered Cases by WHO Region', fontsize=16)
plt.xlabel('WHO Region', fontsize=14)
plt.ylabel('Total Recovered Cases', fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```
